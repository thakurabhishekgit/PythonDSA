```python
class Node:
    def __init__(self , data):
        self.data = data
        self.next = None
class LL:
    def __init__(self):
        self.head = None
    def insertAtBegin(self , data):
        newNode = Node(data)
        newNode.next = self.head
        self.head = newNode
    def insertAtEnd(self , data):
        newNode = Node(data)
        if self.head is None:
            self.head = newNode
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = newNode

    def printLL(self):
        temp = self.head
        while temp:
            print(temp.data , end = " -> ")
            temp = temp.next
        print("None")


l = LL()
l.insertAtEnd(10)
l.insertAtEnd(20)
l.insertAtEnd(30)
l.printLL()
l.insertAtBegin(5)
l.printLL()
```
