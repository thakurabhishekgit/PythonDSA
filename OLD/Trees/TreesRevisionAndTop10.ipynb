```python
from collections import deque
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def insert(root, key):
    if root is None:
        return Node(key)
    if key < root.val:
        root.left = insert(root.left, key)
    else:
        root.right = insert(root.right, key)
    return root


def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=' ')
        inorder(root.right)

def preorder(root):
    if root:
        print(root.val, end=' ')
        preorder(root.left)
        preorder(root.right)        

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=' ')

def BFS(root):
    if not root: return 
    q = deque([root])
    while q:
        node = q.popleft()
        print(node.val, end=' ')
        if node.left: q.append(node.left)
        if node.right: q.append(node.right)

def height(root):
    if not root:
        return 0
    return 1 + max(height(root.left), height(root.right))


def DFS(root):
    if not root: return
    stack = [root]
    while stack:
        node = stack.pop()
        print(node.val, end=' ')
        if node.right: stack.append(node.right)
        if node.left: stack.append(node.left)
        
def search(root , key):
    if root is None or root.val == key:
        return root.val
    if key < root.val:
        return search(root.left, key)
    return search(root.right, key)        

def balance(root):
    if root==None:
        return 0
    lh=height(root.left)
    rh=height(root.right)
    if abs(height(root.left)-height(root.right))<=1:
        print("balance")
    else:                                                                                 #   10
        print("not balance")  
bst  = Node(10)
insert(bst, 5)
insert(bst, 15)
insert(bst, 2)
insert(bst, 7)

inorder(bst)
print()

BFS(bst)

print()

DFS(bst)

search(bst , 2)
print()
balance(bst)
print("-")

print(height(bst))
print()
```
