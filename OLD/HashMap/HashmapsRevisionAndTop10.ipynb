✅ Concept:

Stores key-value pairs (HashMap) or unique elements (HashSet).

O(1) average for search/insert/delete.

⚡ Key Patterns:

Counting frequency of elements (majority element, anagrams)

HashMap + prefix sum (subarray with 0 sum)

Using map to avoid nested loops


| Question                      | Concept                                |
| ----------------------------- | -------------------------------------- |
| Two Sum                       | Find two numbers that add up to target |
| Subarray Sum = 0              | Prefix sum + HashMap                   |
| First Non-Repeating Character | Frequency map                          |
| Group Anagrams                | HashMap + Sorting                      |
| Longest Consecutive Sequence  | HashSet                                |
