from collections import deque

class MyStack:

    def __init__(self):
        self.q = deque()

    def push(self, x: int) -> None:
        self.q.append(x)
        for _ in range(len(self.q) - 1):
            self.q.append(self.q.popleft())
        

    def pop(self) -> int:
        return self.q.popleft() if self.q else None
        

    def top(self) -> int:
        return self.q[0] if not self.empty() else None


    def empty(self) -> bool:
        return len(self.q) == 0
      
s = MyStack()
s.push(1)
s.push(2)
print(s.top())  # 2
print(s.pop())  # 2
print(s.top())  # 1
print(s.empty())  # False
