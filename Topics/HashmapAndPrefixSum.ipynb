```python
# 560. Subarray Sum Equals K
# Medium
# Topics
# Companies
# Hint
# Given an array of integers nums and an integer k, return the total number of subarrays whose sum equals to k.

# A subarray is a contiguous non-empty sequence of elements within an array.

 

# Example 1:

# Input: nums = [1,1,1], k = 2
# Output: 2
# Example 2:

# Input: nums = [1,2,3], k = 3
# Output: 2
 

nums = [3 ,9 ,-2,4,1,-7,2,6,-5,8,-3,-7,6,2,1]
k = 5
total = 0
count = 0
hashmap = {0 : 1}
for num in nums:
    total += num
    if total - k in hashmap:
        count += hashmap[total - k]
    if total in hashmap:
        hashmap[total] += 1
    else:
        hashmap[total] = 1
print(count)
```

```python
# 974. Subarray Sums Divisible by K
# Medium
# Topics
# Companies
# Given an integer array nums and an integer k, return the number of non-empty subarrays that have a sum divisible by k.

# A subarray is a contiguous part of an array.

 

# Example 1:

# Input: nums = [4,5,0,-2,-3,1], k = 5
# Output: 7
# Explanation: There are 7 subarrays with a sum divisible by k = 5:
# [4, 5, 0, -2, -3, 1], [5], [5, 0], [5, 0, -2, -3], [0], [0, -2, -3], [-2, -3]
# Example 2:

# Input: nums = [5], k = 9
# Output: 0
 

# Constraints:

# 1 <= nums.length <= 3 * 104
# -104 <= nums[i] <= 104
# 2 <= k <= 104







# ✅ Step-by-Step Dry Run:
# We’ll maintain:

# prefix_sum: running total

# mod: (prefix_sum % k + k) % k

# mod_map: how many times each mod has occurred

# count: total valid subarrays

# 🔹 Initial:
# ini
# Copy
# Edit
# prefix_sum = 0
# mod_map = {0: 1}
# count = 0
# ▶️ Iteration 1: num = 4
# makefile
# Copy
# Edit
# prefix_sum = 0 + 4 = 4
# mod = (4 % 5 + 5) % 5 = 4
# mod_map[4] = 0 → count += 0 → count = 0
# mod_map[4] += 1 → mod_map = {0:1, 4:1}
# ▶️ Iteration 2: num = 5
# makefile
# Copy
# Edit
# prefix_sum = 4 + 5 = 9
# mod = (9 % 5 + 5) % 5 = 4
# mod_map[4] = 1 → count += 1 → count = 1
# mod_map[4] += 1 → mod_map = {0:1, 4:2}
# ➡️ Why count += 1?
# Because we've seen mod = 4 before → subarray [5] (index 1) is divisible by 5.

# ▶️ Iteration 3: num = 0
# makefile
# Copy
# Edit
# prefix_sum = 9 + 0 = 9
# mod = 4 (same as before)
# mod_map[4] = 2 → count += 2 → count = 3
# mod_map[4] += 1 → mod_map = {0:1, 4:3}
# ➡️ Subarrays: [0], [5, 0]

# ▶️ Iteration 4: num = -2
# makefile
# Copy
# Edit
# prefix_sum = 9 + (-2) = 7
# mod = (7 % 5 + 5) % 5 = (2 + 5) % 5 = 2
# mod_map[2] = 0 → count += 0 → count = 3
# mod_map[2] += 1 → mod_map = {0:1, 4:3, 2:1}
# ▶️ Iteration 5: num = -3
# makefile
# Copy
# Edit
# prefix_sum = 7 + (-3) = 4
# mod = (4 % 5 + 5) % 5 = 4
# mod_map[4] = 3 → count += 3 → count = 6
# mod_map[4] += 1 → mod_map = {0:1, 4:4, 2:1}
# ➡️ Subarrays: [-3], [0, -2, -3], [5, 0, -2, -3]

# ▶️ Iteration 6: num = 1
# makefile
# Copy
# Edit
# prefix_sum = 4 + 1 = 5
# mod = 0
# mod_map[0] = 1 → count += 1 → count = 7
# mod_map[0] += 1 → mod_map = {0:2, 4:4, 2:1}
# ➡️ Subarray: [4, 5, 0, -2, -3, 1]

# ✅ Final Result:
# ini
# Copy
# Edit
# count = 7
# ✅ Summary of Valid Subarrays:
# [5]

# [5, 0]

# [0]

# [0, -2, -3]

# [-3]

# [5, 0, -2, -3]

# [4, 5, 0, -2, -3, 1]



nums = [4,5,0,-2,-3,1]
k = 5
def subarraysDivByK(nums, k):
    from collections import defaultdict

    count = 0
    prefix_sum = 0
    mod_map = defaultdict(int)
    mod_map[0] = 1 

    for num in nums:
        prefix_sum += num
        mod = (prefix_sum % k + k) % k 

        count += mod_map[mod]
        mod_map[mod] += 1

    return count

print(subarraysDivByK(nums,k))
```

```python
# # 523. Continuous Subarray Sum
# # Solved
# # Medium

# # Given an integer array nums and an integer k, return true if nums has a good subarray or false otherwise.

# # A good subarray is a subarray where:

# # its length is at least two, and
# # the sum of the elements of the subarray is a multiple of k.
# # Note that:

# # A subarray is a contiguous part of the array.
# # An integer x is a multiple of k if there exists an integer n such that x = n * k. 0 is always a multiple of k.
 

# # Example 1:

# # Input: nums = [23,2,4,6,7], k = 6
# # Output: true
# # Explanation: [2, 4] is a continuous subarray of size 2 whose elements sum up to 6.
# # Example 2:

# # Input: nums = [23,2,6,4,7], k = 6
# # Output: true
# # Explanation: [23, 2, 6, 4, 7] is an continuous subarray of size 5 whose elements sum up to 42.
# # 42 is a multiple of 6 because 42 = 7 * 6 and 7 is an integer.
# # Example 3:

# # Input: nums = [23,2,6,4,7], k = 13
# # Output: false
 

# # Constraints:

# # 1 <= nums.length <= 105
# # 0 <= nums[i] <= 109
# # 0 <= sum(nums[i]) <= 231 - 1
# # 1 <= k <= 231 - 1


# Initialize:

# prefix_sum = 0

# mod_map = {0: -1} ➝ to handle subarrays starting from index 0

# Loop through the array:

# Add current number to prefix sum

# Compute mod = prefix_sum % k

# If mod has been seen before at index prev_index, and current_index - prev_index >= 2, we return True

# If not seen, store mod in hashmap with current index

# If we complete the loop, return False


# | i | num | prefix\_sum | mod = sum % 6 | mod\_map            | i - mod\_map\[mod] | Result          |
# | - | --- | ----------- | ------------- | ------------------- | ------------------ | --------------- |
# | 0 | 23  | 23          | 5             | {0: -1, 5: 0}       | n/a                |                 |
# | 1 | 2   | 25          | 1             | {0: -1, 5: 0, 1: 1} | n/a                |                 |
# | 2 | 6   | 31          | 1             | (seen at index 1)   | 2 - 1 = 1 ❌        |                 |
# | 3 | 4   | 35          | 5             | (seen at index 0)   | 3 - 0 = 3 ✅        | Return **True** |

from typing import List
class Solution:
    def checkSubarraySum(self, nums: List[int], k: int) -> bool:
        prefix_sum = 0
        mod_map = {0 : -1}
        for i ,num in enumerate(nums):
            prefix_sum += num
            mod = prefix_sum % k
            if mod in mod_map:
                if i - mod_map[mod] > 1:
                    return True
            else:
                mod_map[mod] = i
        return False                
s = Solution()
s.checkSubarraySum([23,2,4,6,7] , 6)
```

```python
# # 1248. Count Number of Nice Subarrays
# # Solved
# # Medium
# # Topics
# # Companies
# # Hint
# # Given an array of integers nums and an integer k. A continuous subarray is called nice if there are k odd numbers on it.

# # Return the number of nice sub-arrays.

 

# # Example 1:

# # Input: nums = [1,1,2,1,1], k = 3
# # Output: 2
# # Explanation: The only sub-arrays with 3 odd numbers are [1,1,2,1] and [1,2,1,1].
# # Example 2:

# # Input: nums = [2,4,6], k = 1
# # Output: 0
# # Explanation: There are no odd numbers in the array.
# # Example 3:

# # Input: nums = [2,2,2,1,2,2,1,2,2,2], k = 2
# # Output: 16
 

# # Constraints:

# # 1 <= nums.length <= 50000
# # 1 <= nums[i] <= 10^5
# # 1 <= k <= nums.length
# | i | num | odd? | prefix\_sum | prefix\_sum - k | freq\[...] used | freq map update | count |
# | - | --- | ---- | ----------- | --------------- | --------------- | --------------- | ----- |
# | 0 | 2   | 0    | 0           | -2              | 0               | {0:2}           | 0     |
# | 1 | 2   | 0    | 0           | -2              | 0               | {0:3}           | 0     |
# | 2 | 2   | 0    | 0           | -2              | 0               | {0:4}           | 0     |
# | 3 | 1   | 1    | 1           | -1              | 0               | {0:4,1:1}       | 0     |
# | 4 | 2   | 0    | 1           | -1              | 0               | {0:4,1:2}       | 0     |
# | 5 | 2   | 0    | 1           | -1              | 0               | {0:4,1:3}       | 0     |
# | 6 | 1   | 1    | 2           | 0               | 4               | {0:4,1:3,2:1}   | 4     |
# | 7 | 2   | 0    | 2           | 0               | 4               | {0:4,1:3,2:2}   | 8     |
# | 8 | 2   | 0    | 2           | 0               | 4               | {0:4,1:3,2:3}   | 12    |
# | 9 | 2   | 0    | 2           | 0               | 4               | {0:4,1:3,2:4}   | 16    |

class Solution:
    def numberOfSubarrays(self, nums: List[int], k: int) -> int:
        count = 0
        prefix_sum = 0
        freq = {0: 1}  # stores frequency of prefix_sum values

        for num in nums:
            if num % 2 == 1:
                prefix_sum += 1

            count += freq.get(prefix_sum - k, 0)
            freq[prefix_sum] = freq.get(prefix_sum, 0) + 1

        return count
s = Solution()
s.numberOfSubarrays([2,2,2,1,2,2,1,2,2,2] , 2)    

```
