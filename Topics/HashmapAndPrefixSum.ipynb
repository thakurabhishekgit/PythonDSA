```python
# 560. Subarray Sum Equals K
# Medium
# Topics
# Companies
# Hint
# Given an array of integers nums and an integer k, return the total number of subarrays whose sum equals to k.

# A subarray is a contiguous non-empty sequence of elements within an array.

 

# Example 1:

# Input: nums = [1,1,1], k = 2
# Output: 2
# Example 2:

# Input: nums = [1,2,3], k = 3
# Output: 2
 

nums = [3 ,9 ,-2,4,1,-7,2,6,-5,8,-3,-7,6,2,1]
k = 5
total = 0
count = 0
hashmap = {0 : 1}
for num in nums:
    total += num
    if total - k in hashmap:
        count += hashmap[total - k]
    if total in hashmap:
        hashmap[total] += 1
    else:
        hashmap[total] = 1
print(count)
```

```python
# 974. Subarray Sums Divisible by K
# Medium
# Topics
# Companies
# Given an integer array nums and an integer k, return the number of non-empty subarrays that have a sum divisible by k.

# A subarray is a contiguous part of an array.

 

# Example 1:

# Input: nums = [4,5,0,-2,-3,1], k = 5
# Output: 7
# Explanation: There are 7 subarrays with a sum divisible by k = 5:
# [4, 5, 0, -2, -3, 1], [5], [5, 0], [5, 0, -2, -3], [0], [0, -2, -3], [-2, -3]
# Example 2:

# Input: nums = [5], k = 9
# Output: 0
 

# Constraints:

# 1 <= nums.length <= 3 * 104
# -104 <= nums[i] <= 104
# 2 <= k <= 104







# âœ… Step-by-Step Dry Run:
# Weâ€™ll maintain:

# prefix_sum: running total

# mod: (prefix_sum % k + k) % k

# mod_map: how many times each mod has occurred

# count: total valid subarrays

# ðŸ”¹ Initial:
# ini
# Copy
# Edit
# prefix_sum = 0
# mod_map = {0: 1}
# count = 0
# â–¶ï¸ Iteration 1: num = 4
# makefile
# Copy
# Edit
# prefix_sum = 0 + 4 = 4
# mod = (4 % 5 + 5) % 5 = 4
# mod_map[4] = 0 â†’ count += 0 â†’ count = 0
# mod_map[4] += 1 â†’ mod_map = {0:1, 4:1}
# â–¶ï¸ Iteration 2: num = 5
# makefile
# Copy
# Edit
# prefix_sum = 4 + 5 = 9
# mod = (9 % 5 + 5) % 5 = 4
# mod_map[4] = 1 â†’ count += 1 â†’ count = 1
# mod_map[4] += 1 â†’ mod_map = {0:1, 4:2}
# âž¡ï¸ Why count += 1?
# Because we've seen mod = 4 before â†’ subarray [5] (index 1) is divisible by 5.

# â–¶ï¸ Iteration 3: num = 0
# makefile
# Copy
# Edit
# prefix_sum = 9 + 0 = 9
# mod = 4 (same as before)
# mod_map[4] = 2 â†’ count += 2 â†’ count = 3
# mod_map[4] += 1 â†’ mod_map = {0:1, 4:3}
# âž¡ï¸ Subarrays: [0], [5, 0]

# â–¶ï¸ Iteration 4: num = -2
# makefile
# Copy
# Edit
# prefix_sum = 9 + (-2) = 7
# mod = (7 % 5 + 5) % 5 = (2 + 5) % 5 = 2
# mod_map[2] = 0 â†’ count += 0 â†’ count = 3
# mod_map[2] += 1 â†’ mod_map = {0:1, 4:3, 2:1}
# â–¶ï¸ Iteration 5: num = -3
# makefile
# Copy
# Edit
# prefix_sum = 7 + (-3) = 4
# mod = (4 % 5 + 5) % 5 = 4
# mod_map[4] = 3 â†’ count += 3 â†’ count = 6
# mod_map[4] += 1 â†’ mod_map = {0:1, 4:4, 2:1}
# âž¡ï¸ Subarrays: [-3], [0, -2, -3], [5, 0, -2, -3]

# â–¶ï¸ Iteration 6: num = 1
# makefile
# Copy
# Edit
# prefix_sum = 4 + 1 = 5
# mod = 0
# mod_map[0] = 1 â†’ count += 1 â†’ count = 7
# mod_map[0] += 1 â†’ mod_map = {0:2, 4:4, 2:1}
# âž¡ï¸ Subarray: [4, 5, 0, -2, -3, 1]

# âœ… Final Result:
# ini
# Copy
# Edit
# count = 7
# âœ… Summary of Valid Subarrays:
# [5]

# [5, 0]

# [0]

# [0, -2, -3]

# [-3]

# [5, 0, -2, -3]

# [4, 5, 0, -2, -3, 1]



nums = [4,5,0,-2,-3,1]
k = 5
def subarraysDivByK(nums, k):
    from collections import defaultdict

    count = 0
    prefix_sum = 0
    mod_map = defaultdict(int)
    mod_map[0] = 1 

    for num in nums:
        prefix_sum += num
        mod = (prefix_sum % k + k) % k 

        count += mod_map[mod]
        mod_map[mod] += 1

    return count

print(subarraysDivByK(nums,k))
```
