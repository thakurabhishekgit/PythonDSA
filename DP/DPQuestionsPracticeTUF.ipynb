```python
def frog_k(n, k, height, dp):
    if n == 0:
        return 0

    if dp[n] != -1:
        return dp[n]

    min_cost = float('inf')

    for j in range(1, k + 1):
        if n - j >= 0:
            cost = frog_k(n - j, k, height, dp) + abs(height[n] - height[n - j])
            min_cost = min(min_cost, cost)

    dp[n] = min_cost
    return dp[n]


height = [10, 20, 10, 30, 20]
n = len(height) - 1
k = 3
dp = [-1] * (n + 1)

print(frog_k(n, k, height, dp))
```

```python
# Triamgle
from typing import List
class Solution:
    def minimumTotal(self, triangle: List[List[int]]) -> int:
        dp = triangle[-1][:]  
        for row in range(len(triangle) - 2, -1, -1):
            for i in range(len(triangle[row])):
                dp[i] = triangle[row][i] + min(dp[i], dp[i + 1])
        return dp[0]

s = Solution()
s.minimumTotal([[2],
                [3,4],
                [6,5,7],
                [4,1,8,3]]
                )
        
```

```python
from typing import List
class Solution:
    def minFallingPathSum(self, matrix: List[List[int]]) -> int:
        n = len(matrix)
        dp = [[0 for _ in range(n)] for _ in range(n)]

        for j in range(n):
            dp[0][j] = matrix[0][j]
       
        for i in range(1, n):
            for j in range(n):
                up = dp[i - 1][j]
                left_diag = dp[i - 1][j - 1] if j > 0 else float('inf')
                right_diag = dp[i - 1][j + 1] if j < n - 1 else float('inf')

                dp[i][j] = matrix[i][j] + min(up, left_diag, right_diag)

        return min(dp[n - 1])
s = Solution()
s.minFallingPathSum([[2,1,3],[6,5,4],[7,8,9]])
```

```python
from typing import List
#741. Cherry Pickup
class Solution(object):
    def cherryPickup(self, grid):
        n = len(grid)

        dp = [[float('-inf')]*n for _ in range(n)]
        dp[0][0] = 0

        for k in range(2*(n-1) + 1):

            newDp = [[float('-inf')]*n for _ in range(n)]
            for x1 in range(n):
                y1 = k - x1
                if not(0 <= y1 < n):
                    continue

                for x2 in range(n):
                    y2 = k - x2
                    if not(0 <= y2 < n):
                        continue
                    
                    if grid[x1][y1] == -1 or grid[x2][y2] == -1:
                        continue
                    
                    newDp[x1][x2] = dp[x1][x2]
                    if x1 > 0:
                        newDp[x1][x2] = max(newDp[x1][x2], dp[x1-1][x2])
                    if x2 > 0:
                        newDp[x1][x2] = max(newDp[x1][x2], dp[x1][x2-1])
                    if x1 >0 and x2 > 0:
                        newDp[x1][x2] = max(newDp[x1][x2], dp[x1-1][x2-1])
                    
                    newDp[x1][x2] += grid[x1][y1]
                    if x1 != x2:
                        newDp[x1][x2] += grid[x2][y2]
            dp = newDp

        return max(0, dp[n-1][n-1])
    
s = Solution()
s.cherryPickup([[0,1,-1],[1,0,-1],[1,1,1]])
```

```python
from typing import List
class Solution:
    def canPartition(self, nums: List[int]) -> bool:
        total = sum(nums)
        if total % 2 == 1:
            return False
        
        target = total // 2

        dp = [False] * (target + 1)
        dp[0] = True

        for n in nums:
            for i in range(len(dp) - 1, n - 1, -1):
                if dp[i]: continue
                if dp[i-n]: dp[i] = True
                if dp[-1]: return True
        
        return False
    
s = Solution()
s.canPartition([1,2,3,5])
```

```python

# 1200. Minimum Absolute Difference

class Solution:
    def minimumAbsDifference(self, arr: List[int]) -> List[List[int]]:

        arr.sort()
        ans = []
        minn = float('inf')

        for i in range(len(arr) - 1):
            val = abs(arr[i] - arr[i + 1])
            minn = min (minn , val)

        for i in range(len(arr) - 1):
            val = abs(arr[i] - arr[i + 1])
            if val == minn:
                ans.append([arr[i], arr[i + 1]])

        return ans        


s = Solution()
s.minimumAbsDifference([40,11,26,27,-20])
        
```
