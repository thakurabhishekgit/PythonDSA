```python
def fibo(n , dp):
    if n <= 1:
        return n
    if dp[n] != -1:
        return dp[n]
    
    dp[n] =  fibo(n - 1 , dp) + fibo(n - 2 , dp)
    return dp[n]

n = 5
dp = [-1] * (n + 1)
print(fibo(n , dp))
print(dp)



```

```python
def climb(n, dp):
    if n <= 1:
        return 1

    if dp[n] != -1:
        return dp[n]

    dp[n] = climb(n - 1, dp) + climb(n - 2, dp)
    return dp[n]

n = 3
dp = [-1] * (n + 1)
print(climb(n, dp))
```
