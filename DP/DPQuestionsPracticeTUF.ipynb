```python
def fibo(n , dp):
    if n <= 1:
        return n
    if dp[n] != -1:
        return dp[n]
    
    dp[n] =  fibo(n - 1 , dp) + fibo(n - 2 , dp)
    return dp[n]

n = 5
dp = [-1] * (n + 1)
print(fibo(n , dp))
print(dp)



```

```python
def climb(n, dp):
    if n <= 1:
        return 1

    if dp[n] != -1:
        return dp[n]

    dp[n] = climb(n - 1, dp) + climb(n - 2, dp)
    return dp[n]

n = 3
dp = [-1] * (n + 1)
print(climb(n, dp))
```

```python
height = [10, 20, 10, 30, 20]
n = len(height)

dp = [0] * n  

dp[0] = 0   

for i in range(1, n):
    left = dp[i - 1] + abs(height[i] - height[i - 1])
    right = float('inf')

    if i > 1:
        right = dp[i - 2] + abs(height[i] - height[i - 2])

    dp[i] = min(left, right)

print(dp[n - 1])
```

```python
def frog_k(n, k, height, dp):
    if n == 0:
        return 0

    if dp[n] != -1:
        return dp[n]

    min_cost = float('inf')

    for j in range(1, k + 1):
        if n - j >= 0:
            cost = frog_k(n - j, k, height, dp) + abs(height[n] - height[n - j])
            min_cost = min(min_cost, cost)

    dp[n] = min_cost
    return dp[n]


height = [10, 20, 10, 30, 20]
n = len(height) - 1
k = 3
dp = [-1] * (n + 1)

print(frog_k(n, k, height, dp))
```

```python
def maxSum(n, dp):
    if n == 0:
        return nums[0]

    if n < 0:
        return 0

    if dp[n] != -1:
        return dp[n]

    pick = nums[n] + maxSum(n - 2, dp)
    notPick = maxSum(n - 1, dp)

    dp[n] = max(pick, notPick)
    return dp[n]


nums = [2, 1, 4, 9]
n = len(nums) - 1
dp = [-1] * (n + 1)

print(maxSum(n, dp))
```

```python
# tabulation for max sum of non adjacent elements
nums = [2,1,4,9]
n = len(nums)
dp = [0] * n
for i in range(n):
    take = nums[i] + (dp[i - 2] if i - 2 >= 0 else 0)
    notTake = 0 + (dp[i - 1] if i - 1 >= 0 else 0)
    dp[i] = max(take , notTake)

print(dp[n - 1])    
```
