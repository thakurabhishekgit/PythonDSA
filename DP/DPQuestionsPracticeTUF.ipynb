```python
def fibo(n , dp):
    if n <= 1:
        return n
    if dp[n] != -1:
        return dp[n]
    
    dp[n] =  fibo(n - 1 , dp) + fibo(n - 2 , dp)
    return dp[n]

n = 5
dp = [-1] * (n + 1)
print(fibo(n , dp))
print(dp)



```

```python
def climb(n, dp):
    if n <= 1:
        return 1

    if dp[n] != -1:
        return dp[n]

    dp[n] = climb(n - 1, dp) + climb(n - 2, dp)
    return dp[n]

n = 3
dp = [-1] * (n + 1)
print(climb(n, dp))
```

```python
def frog_k(n, k, height, dp):
    if n == 0:
        return 0

    if dp[n] != -1:
        return dp[n]

    min_cost = float('inf')

    for j in range(1, k + 1):
        if n - j >= 0:
            cost = frog_k(n - j, k, height, dp) + abs(height[n] - height[n - j])
            min_cost = min(min_cost, cost)

    dp[n] = min_cost
    return dp[n]


height = [10, 20, 10, 30, 20]
n = len(height) - 1
k = 3
dp = [-1] * (n + 1)

print(frog_k(n, k, height, dp))
```

```python
def maxSum(n, dp):
    if n == 0:
        return nums[0]

    if n < 0:
        return 0

    if dp[n] != -1:
        return dp[n]

    pick = nums[n] + maxSum(n - 2, dp)
    notPick = maxSum(n - 1, dp)

    dp[n] = max(pick, notPick)
    return dp[n]


nums = [2, 1, 4, 9]
n = len(nums) - 1
dp = [-1] * (n + 1)

print(maxSum(n, dp))
```

```python
# grid unique paths using tabulation
def gridUnique(n , m):  
    dp = [[0 for _ in range(m + 1)] for _ in range(n + 1)]
    dp[1][1] = 1

    for i in range(1 , n + 1):
        for j in range(1 , m + 1):
            if i == 1 and j == 1:
                continue
            up = dp[i - 1][j]
            left = dp[i][j - 1]
            dp[i][j] = up + left
    return dp[n][m]

n = 3
m = 3
print(gridUnique(n , m))
```
